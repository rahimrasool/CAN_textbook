### Matplotlib and its structure 

According to its documentation, Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

Within matplotlib, exists pyplot, which offers numerous functions and interface for plotting. We are going to import oth matplotlib and pyplot and use it ahead for creating figures and plots.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

The following image explains the anatomy of a Matplotlib Figure. Understanding this is a critical step towards learning how matplotlib works and how to create an customize plots.

<img src="https://matplotlib.org/stable/_images/sphx_glr_anatomy_001.png" width="400"/>

The base container in matplotlib is the Figure. The figure holds all elements of the plot. Think of it as a canvas on which there can be multiple plots. The figure can hold multiple child elements such as plots, and it also tracks all these child elements within it. To relate it to the image above, the Figure is the container that holds the plot and other elements that we see. In image above is is called `plt.figure` which refers to a pyplot figure.

As mentioned, figure holds multiple elements which are referred to as Artists. Axes is one of those Artists that forms a region of a plot which can display 2 or 3 dimensional data in form of a plot. Therefore, there can be multiple Axes (or plot regions) on a Figure. Each Axes also has a title, an x-label, and a y-label. Notice that Axes in the image above is called as `fig.subplots` which means that it is an object within the `plt.figure` class.

Within Axes objects there are 2 or 3 Axis(Different from Axes). Axis is also an Artist object. The Axis sets the scale and limits. It also set ticks and tick labels. Notice the y Axis and x Axis in the image above which will be referred to as `ax.xaxis` and `ax.yaxis`, respectively, since both are objects of the Axes class. Also notice that the ticks and labels are part of the Axis in the image above.

Every other Artist in the image above, such as the line, markers, grid, legend etc, are part of the Axes class.

### Object-Oriented Interface

It is essential to understand the two styles (application interface) to use matplotlib. We will start with the Object-oriented interface, also called the Axes interface and then move on to the pyplot interface. The Object-oriented is an explicit interface and is the one that we highly recommend you to primarily work on in order to avoid errors and confusion.

The object-oriented interface build visuals in a ste-by-step stuctured manner. As described above, it requires to define the figure, the set of Axes from that figure and the other Artists object on those Axes and figure. Since we are explicitly defining each Artist object including Figure and Axes, it enables us to customize these objects.

We are going to work on an example of using this technique to create a few visualization below.

In the the first interface below, we create an instance of the Figure class called `fig` and then we use the `.subplots()` on fig to create Axes within the Figure. The Axes are assigned to variable `ax`. We then use Axes' `.plot()` method to create the line plot over it. As you can notice below, the plot takes a list of x-axis valuea and the corresponding y-axis values to create a line plot as shown below.

```python
fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

We will create another Figure with 2 Axes now, displaying a line plot and scatter plot (with markers). The subplot takes the number of rows and columns to display on the figure, respectively. This creates 2 Axes objects, where each object can be referred using integer index as shown below. 

```python
fig = plt.figure()
ax = fig.subplots(1, 2)
ax[0].plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
ax[1].scatter([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

We can add more Artists object to each Axes.

```python
fig = plt.figure()
ax = fig.subplots(1, 2)

ax[0].plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
ax[0].set_ylabel("distance (m)")
ax[0].set_xlabel("time")

ax[1].scatter([1, 2, 3, 4], [0, 0.5, 1, 0.2])
ax[1].xaxis.grid()
ax[1].yaxis.grid()
```

### Pyplot Interface

The Pyplot interface is a state-based interface. State-based means that it creates a Figure and its Axes, tracks the state of these objecs and modifies them with various function calls on that state. 

This can be convenient, particularly when doing interactive work or simple scripts. However, since it only tracks the state of the last figure and axes, it becomes confusing to modify a specific Axes when required. Therefore, it is preferred to use the object-oriented approach. The reason to introduce and understand this concept is to familiarise you with the difference and get an idea when you review snippets of code using this style of plotting.

The command below creates the same plot as we create above, but it also creates the figure and Axes along.

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

We will modify the plot we create by applying methods on the current state.

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid(True)
```

### Understanding basic plots

We will go though a few simple exaples of common plots that we can create using matplotlib's pyplot. In each of the examples, we will first create a figure, create an axis on that figure and then specify the plot on that axis. 

In the first example below, we will use the `.plot()` method on the Axes to create a simple line plot. We provide a list of x and y coordinates in a list. The format is:

`ax.plot(x, y, ...)`

```python
fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

Using a similar approach as above, we will use the `.scatter()` method on the Axes object to create a scatter plot using the same coordinates provided.

```python
fig = plt.figure()
ax = fig.subplots()
ax.scatter([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

Using `.bar()`, we will use the same list to create a bar plot. In this case the first list specifies the x axis placement of each bar, whereas, the second list is the height of each bar. The format is:

`ax.bar(x, height, ...)`

```python
fig = plt.figure()
ax = fig.subplots()
ax.bar([1, 2, 3, 4], [0.3, 0.5, 1, 0.2])
```

### Advanced Plot
